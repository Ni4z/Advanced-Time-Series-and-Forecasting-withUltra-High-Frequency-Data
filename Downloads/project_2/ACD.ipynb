```r
X$Time <- format(as.POSIXct(X$Time, format = "%H:%M:%OS"), "%H:%M:%OS")
```

```r
durations <- as.numeric(diff(as.POSIXct(X$Time, format = "%H:%M:%OS")), units = "secs")
```

```r

acd_model <- acdFit(durations, dist = "exponential",
                    optimFnc = "nlminb",
                    startPara = c(0.1, 0.1, 0.1),
                    control = list(trace = 1))
```

```r
summary(acd_model)
```

```r
intensities <- 1 / durations
```

```r
plot_data <- data.frame(Index = seq_along(durations), Durations = durations)
```

```r
plot(intensities, type = "l", col = "blue", xlab = "Trade", ylab = "Intensity",
     main = "Estimated Trading Intensities Over Time")
```

```r
plot(intensities, type = "l", col = "blue", xlab = "Trade", ylab = "Intensity",
     main = "Estimated Trading Intensities Over Time", xlim = c(1, length(intensities)))
```

```r
par(mar = c(5, 4, 4, 2) + 0.1)
plot(intensities, type = "l", col = "blue", xlab = "Trade", ylab = "Intensity",
     main = "Estimated Trading Intensities Over Time", xlim = c(1, length(intensities)))
```

```r
# Create a subset of data for a less dense plot
thin_factor <- 10  # Adjust thin_factor as needed to thin out the points
thin_intensities <- intensities[seq(1, length(intensities), by = thin_factor)]
```

```r


# Plot the thinned out data
plot(thin_intensities, type = "l", col = "black", xlab = "Trade",lwd = 3, ylab = "Intensity",
     main = "Estimated Trading Intensities Over Time (Thinned)")
```
